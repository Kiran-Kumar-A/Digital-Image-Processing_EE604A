```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('iitk.jpg')
w,h,ch =img.shape
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.bilateralFilter(img, 12, 85, 65)
img_out = np.zeros_like(img)
threshold, img_out1 = cv2.threshold(img,110, 255,cv2.THRESH_OTSU)
for i in range(w):
  for j in range(h):
    if(img[i,j] > (int(threshold/1.4))):
      img_out[i,j]=0
    else:
      img_out[i,j]=255
plt.figure(figsize=[20,20], dpi=100)
plt.subplot(1, 2, 1)
plt.imshow(img,cmap='gray');plt.title("Original")
plt.subplot(1, 2, 2)
plt.imshow(img_out,cmap='gray');plt.title("segmented")
cv2.imwrite('fordrone.jpg', img_out, [int(cv2.IMWRITE_JPEG_QUALITY), 100])

```
